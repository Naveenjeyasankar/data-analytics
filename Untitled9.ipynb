```python
import matplotlib as mpl
from matplotlib import pyplot as plt
import numpy as np
```

```python
plt.style.use('classic')
%matplotlib inline
x=np.linspace(0,10,100)
fig=plt.figure()
plt.plot(x,np.sin(x),'-')
plt.plot(x,np.cos(x),'--')
fig.savefig('my_name.png')
plt.show()

```

```python
fig,ax=plt.subplots(2)
ax[0].plot(x,np.sin(x))
ax[1].plot(x,np.cos(x))
plt.show()
```
